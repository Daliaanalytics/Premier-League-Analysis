```python
df['Nation']= df['Nation'].str.split(' ').str[-1]
df['Nation'].head(10)
```

```python
df['Pos']= df['Pos'].str.split(',').str[0]
df['Pos'].value_counts()
```

```python
top_scorers= df[['Player', 'Team', 'Gls']].sort_values('Gls', ascending= False).head(10)
top_scorers
```

```python
top_scorer= df[['Player', 'Team', 'Gls']].nlargest(10, 'Gls')
top_scorer
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_scorer.reset_index(drop= True), x='Gls', y='Player', palette= 'viridis')
plt.title('Top 10 Scorers - Premier League')
plt.xlabel('Goals')
plt.ylabel('player')
plt.tight_layout()
plt.show()
```

```python
top_assists= df[['Player', 'Team', 'Ast']].nlargest(10, 'Ast')
top_assists
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_assists.reset_index(drop=True), x='Ast', y='Player', palette= 'magma' )
plt.title('Top 10 Assists - Premier League')
plt.xlabel('Assists')
plt.ylabel('Player')
plt.tight_layout()
plt.show()
```

```python
top_ga=df[['Player','Team','Gls','Ast','G+A']].nlargest(10,'G+A')
top_ga
```

```python
df.head()
```

```python
top_ga_sorted= top_ga.sort_values('G+A', ascending= True)
plt.figure(figsize=(10,6))
plt.barh(top_ga_sorted['Player'], top_ga_sorted['Gls'], color='steelblue', label= 'Goals')
plt.barh(top_ga_sorted['Player'], top_ga_sorted['Ast'], left=top_ga_sorted['Gls'], color='coral', label='Assists'  )
plt.title('Top 10 Goals + Assists - Premier League')
plt.xlabel('G+A')
plt.tight_layout()
plt.show()
```

```python
df['Team'].value_counts()
```

```python
team_goals= df.groupby('Team')['Gls'].sum().sort_values(ascending=False)
team_goals
```

```python
plt.figure(figsize=(12,7))

sns.barplot(x= team_goals.values,y= team_goals.index,hue= team_goals.index, palette='coolwarm' , legend=False)
plt.title('Total goals by Team- Premier League')
plt.xlabel('Goals')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
team_players=df.groupby('Team')['Player'].count().sort_values(ascending= False)
team_players
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x= team_players.values, y=team_players.index, hue=team_players.index, palette= 'Set2', legend=False)
plt.title('Number of Players by Team- Premier League')
plt.xlabel('Players')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
top_nations= df['Nation'].value_counts().head(15)
top_nations
```

```python
plt.figure(figsize=(10,8))
plt.pie(top_nations.values, labels= top_nations.index, autopct='%1.1f%%', startangle= 140)
plt.title('Top 15 Nationalities - Premier League')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Age'], bins=20, kde= True, color='steelblue')
plt.title('Age Distribution - Premier League Players')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
top_mins= df[['Player','Team','Min']].nlargest(10, 'Min')
top_mins
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_mins.reset_index(drop=True), x='Min', y='Player', hue='Player', palette='Blues_r', legend=False)
plt.title('Top 10 Players by Minutes Played - Premier League')
plt.xlabel('Minutes')
plt.ylabel('Player')
plt.tight_layout()
plt.show()
```
